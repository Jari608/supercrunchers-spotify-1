```python
import pandas as pd

import statsmodels.api as sm

from sklearn.model_selection import train_test_split
```

## Averages

```python
def open_data(file, norm):
    df = pd.read_csv(file)
    df.columns

    return df

def norm_data(df_norm):
    
    df_norm[['acousticness', 'danceability', 'duration_ms', 'energy',
       'instrumentalness', 'liveness', 'loudness', 'speechiness',
       'tempo', 'valence', 'popularity_1', 'acousticness1',
       'danceability1', 'duration_ms1', 'energy1', 'instrumentalness1',
       'liveness1', 'loudness1', 'speechiness1', 'tempo1',
       'valence1']] = df_norm[['acousticness', 'danceability', 'duration_ms', 'energy',
       'instrumentalness', 'liveness', 'loudness', 'speechiness',
       'tempo', 'valence', 'popularity_1', 'acousticness1',
       'danceability1', 'duration_ms1', 'energy1', 'instrumentalness1',
       'liveness1', 'loudness1', 'speechiness1', 'tempo1',
       'valence1']].apply(lambda x: x*100/(x).max(), axis=0)
    
    return df_norm
```

```python
def gen_difference_table(df):

    tempo_diff = df.groupby('genre_allmusic')[['tempo', 'tempo1']].mean().diff(axis = 1).reset_index()[['genre_allmusic', 'tempo1']].rename(columns = {'tempo1': 'Difference'})
    print("Mean Tempo: {}".format(tempo_diff[['Difference']].mean().values[0]))
    tempo_diff_ = tempo_diff.style.set_caption("Tempo Difference between Original And Most popular Cover (BPM)")

    popularity_diff = df.groupby('genre_allmusic')[['popularity', 'popularity_1']].mean().diff(axis = 1).reset_index()[['genre_allmusic', 'popularity_1']].rename(columns = {'popularity_1': 'Difference'})
    print("Mean Popularity: {}".format(popularity_diff[['Difference']].mean().values[0]))
    popularity_diff_ = popularity_diff.style.set_caption("Popularity Difference between Original And Most popular Cover")

    acousticness_diff = df.groupby('genre_allmusic')[['acousticness', 'acousticness1']].mean().diff(axis = 1).reset_index()[['genre_allmusic', 'acousticness1']].rename(columns = {'acousticness1': 'Difference'})
    print("Mean Acousticness: {}".format(acousticness_diff[['Difference']].mean().values[0]))
    acousticness_diff_ = acousticness_diff.style.set_caption("acousticness Difference between Original And Most popular Cover")

    danceability_diff = df.groupby('genre_allmusic')[['danceability', 'danceability1']].mean().diff(axis = 1).reset_index()[['genre_allmusic', 'danceability1']].rename(columns = {'danceability1': 'Difference'})
    print("Mean Danceability: {}".format(danceability_diff[['Difference']].mean().values[0]))
    danceability_diff_ = danceability_diff.style.set_caption("danceability Difference between Original And Most popular Cover")

    duration_ms_diff = df.groupby('genre_allmusic')[['duration_ms', 'duration_ms1']].mean().diff(axis = 1).reset_index()[['genre_allmusic', 'duration_ms1']].rename(columns = {'duration_ms1': 'Difference'})
    print("Mean Duration (s): {}".format( int(duration_ms_diff[['Difference']].mean().values[0]) / 1000))
    duration_ms_diff['Difference'] = duration_ms_diff['Difference'].apply(lambda x: x / 1000)
    duration_ms_diff_ = duration_ms_diff.style.set_caption("duration_ms Difference between Original And Most popular Cover (seconds) ")

    energy_diff = df.groupby('genre_allmusic')[['energy', 'energy1']].mean().diff(axis = 1).reset_index()[['genre_allmusic', 'energy1']].rename(columns = {'energy1': 'Difference'})
    print("Mean Energy: {}".format(energy_diff[['Difference']].mean().values[0]))
    energy_diff_ = energy_diff.style.set_caption("energy_diff difference between Original And Most popular Cover")

    instrumentalness_diff = df.groupby('genre_allmusic')[['instrumentalness', 'instrumentalness1']].mean().diff(axis = 1).reset_index()[['genre_allmusic', 'instrumentalness1']].rename(columns = {'instrumentalness1': 'Difference'})
    print("Mean Instrumentalness: {}".format(instrumentalness_diff[['Difference']].mean().values[0]))
    instrumentalness_diff_ = instrumentalness_diff.style.set_caption("instrumentalness difference between Original And Most popular Cover")

    liveness_diff = df.groupby('genre_allmusic')[['liveness', 'liveness1']].mean().diff(axis = 1).reset_index()[['genre_allmusic', 'liveness1']].rename(columns = {'liveness1': 'Difference'})
    print("Mean Liveness: {}".format(liveness_diff[['Difference']].mean().values[0]))
    liveness_diff_ = liveness_diff.style.set_caption("instrumentalness difference between Original And Most popular Cover")

    loudness_diff = df.groupby('genre_allmusic')[['loudness', 'loudness1']].mean().diff(axis = 1).reset_index()[['genre_allmusic', 'loudness1']].rename(columns = {'loudness1': 'Difference'})
    print("Mean Loudness: {}".format(loudness_diff[['Difference']].mean().values[0]))
    loudness_diff_ = loudness_diff.style.set_caption("instrumentalness difference between Original And Most popular Cover")

    speechiness_diff = df.groupby('genre_allmusic')[['speechiness', 'speechiness1']].mean().diff(axis = 1).reset_index()[['genre_allmusic', 'speechiness1']].rename(columns = {'speechiness1': 'Difference'})
    print("Mean Speechiness: {}".format(speechiness_diff[['Difference']].mean().values[0]))
    speechiness_diff_ = speechiness_diff.style.set_caption("instrumentalness difference between Original And Most popular Cover")

    valence_diff = df.groupby('genre_allmusic')[['valence', 'valence1']].mean().diff(axis = 1).reset_index()[['genre_allmusic', 'valence1']].rename(columns = {'valence1': 'Difference'})
    print("Mean Valence: {}".format(valence_diff[['Difference']].mean().values[0]))
    valence_diff_ = valence_diff.style.set_caption("instrumentalness difference between Original And Most popular Cover")

    df_total = tempo_diff.set_index('genre_allmusic').join(energy_diff.set_index('genre_allmusic'), on = 'genre_allmusic', lsuffix = '_Tempo', rsuffix = '_Energy')
    df_total = df_total.rename(columns = {'Difference_Tempo': 'Difference Tempo (BPM)'})
    df_total = df_total.join(duration_ms_diff.set_index('genre_allmusic')).rename(columns = {'Difference': 'Difference Duration (s)'})
    df_total = df_total.join(popularity_diff.set_index('genre_allmusic')).rename(columns = {'Difference': 'Difference Popularity'})
    df_total = df_total.join(acousticness_diff.set_index('genre_allmusic')).rename(columns = {'Difference': 'Difference Acousticness'})
    df_total = df_total.join(danceability_diff.set_index('genre_allmusic')).rename(columns = {'Difference': 'Difference Danceability'})
    df_total = df_total.join(instrumentalness_diff.set_index('genre_allmusic')).rename(columns = {'Difference': 'Difference Instrumentalness'})
    df_total = df_total.join(liveness_diff.set_index('genre_allmusic')).rename(columns = {'Difference': 'Difference Liveness'})
    df_total = df_total.join(loudness_diff.set_index('genre_allmusic')).rename(columns = {'Difference': 'Difference Loudness'})
    df_total = df_total.join(speechiness_diff.set_index('genre_allmusic')).rename(columns = {'Difference': 'Difference Speechiness'})
    df_total = df_total.join(valence_diff.set_index('genre_allmusic')).rename(columns = {'Difference': 'Difference Valence'})

    return df_total

def _color_red_or_green(val):
    color = 'red' if val < 0 else 'green'
    return 'color: %s' % color
```

```python

def rename_features(df):
    
    df['tempo_diff'] = df['tempo'].sub(df['tempo1'], axis = 0)
    df['energy_diff'] = df['energy'].sub(df['energy1'], axis = 0)
    df['duration_ms_diff'] = df['duration_ms'].sub(df['duration_ms1'], axis = 0)
    df['popularity_diff'] = df['popularity'].sub(df['popularity_1'], axis = 0)
    df['acousticness_diff'] = df['acousticness'].sub(df['acousticness1'], axis = 0)
    df['danceability_diff'] = df['danceability'].sub(df['danceability1'], axis = 0)
    df['instrumentalness_diff'] = df['instrumentalness'].sub(df['instrumentalness1'], axis = 0)
    df['liveness_diff'] = df['liveness'].sub(df['liveness1'], axis = 0)
    df['loudness_diff'] = df['loudness'].sub(df['loudness1'], axis = 0)
    df['speechiness_diff'] = df['speechiness'].sub(df['speechiness1'], axis = 0)
    df['valence_diff'] = df['valence'].sub(df['valence1'], axis = 0)

    return df
```

```python
df = open_data('Data\\genres_csv\\final_set_remixes_top1.csv', norm = False)

df = rename_features(df)

X = df.filter(items = ['tempo_diff', 'energy_diff', 'duration_ms_diff', 'acousticness_diff', 'danceability_diff', 'instrumentalness_diff', 'liveness_diff', 'loudness_diff', 'speechiness_diff', 'valence_diff'])
y= df['popularity_diff'].to_list()

X = sm.add_constant(X)

result = sm.OLS(y, X).fit()

print(result.summary())
```
